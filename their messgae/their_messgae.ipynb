## 3. vizuilize our data using TSNE algorithm to show the clusters

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt



# Apply t-SNE for dimensionality reduction
tsne = TSNE(n_components=2, random_state=42)
data1_tsne = tsne.fit_transform(data1)
data2_tsne = tsne.fit_transform(data2)
data3_tsne = tsne.fit_transform(data3)
data4_tsne = tsne.fit_transform(data4)

# Scatter plot for data1
plt.figure(figsize=(8, 6))
plt.scatter(data1_tsne[:, 0], data1_tsne[:, 1], label='Data 1')
plt.title('t-SNE Scatter Plot for Data 1')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot for data2
plt.figure(figsize=(8, 6))
plt.scatter(data2_tsne[:, 0], data2_tsne[:, 1], label='Data 2')
plt.title('t-SNE Scatter Plot for Data 2')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot for data3
plt.figure(figsize=(8, 6))
plt.scatter(data3_tsne[:, 0], data3_tsne[:, 1], label='Data 3')
plt.title('t-SNE Scatter Plot for Data 3')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid(True)
plt.show()

# Scatter plot for data4
plt.figure(figsize=(8, 6))
plt.scatter(data4_tsne[:, 0], data4_tsne[:, 1], label='Data 4')
plt.title('t-SNE Scatter Plot for Data 4')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.grid(True)
plt.show()
```
